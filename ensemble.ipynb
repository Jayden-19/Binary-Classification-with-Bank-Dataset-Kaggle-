```python
import pandas as pd

X_test = pd.read_csv('playground-series-s5e8/test.csv')
```

```python
import joblib
from tensorflow.keras.models import load_model

# Load preprocessor
preprocessor = joblib.load('preprocessor.pkl')

# Load TensorFlow model
model = load_model('tensorflow_model.h5')

# Transform test data
X_test_transformed = preprocessor.transform(X_test)

# Predict probabilities
y_pred_tf = model.predict(X_test_transformed).flatten()  # 1D array
```

```python
pipeline = joblib.load('lightgbm_pipeline.pkl')

# Predict probabilities directly
y_pred_lgb = pipeline.predict_proba(X_test)[:, 1]  # probability of class 1
```

```python
# Example: LightGBM is stronger
y_ensemble = 0.7 * y_pred_lgb + 0.3 * y_pred_tf
```

```python
submission = pd.DataFrame({
    'id':X_test['id'],
    'y':y_ensemble
})

submission.to_csv('submission6.csv', index=False)
```
